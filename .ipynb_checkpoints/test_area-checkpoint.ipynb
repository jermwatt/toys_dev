```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
```

```python
x = np.linspace(0,1,100)
y = np.exp(x)*np.cos(2*math.pi*np.sin(math.pi*x))
plt.plot(x,y)
```

```python
num_pts = 39
noise_level = 0.25
inds = np.random.permutation(len(x))
x_data = x[inds[0:num_pts]]
y_data = y[inds[0:num_pts]] 
y_data = y_data + noise_level*np.random.randn(len(y_data),)
plt.scatter(x_data,y_data)
plt.plot(x,y)
```
