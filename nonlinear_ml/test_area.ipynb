```python
# generate steppy function
x = np.linspace(0,1,100)
num_splits = 4
vs = np.random.permutation(range(10,90))[:num_splits]
vs = [0,vs[0],vs[1],vs[2],vs[3],100]
vs = np.sort(vs)
diffs =[]
for n in range(num_splits+1):
    diffs.append(vs[n+1]-vs[n])

func = []
levels = np.random.randn(num_splits+1,1)
for i in range(num_splits+1):
    step = list(np.repeat(levels[i],diffs[i]))[:]
    [func.append(s) for s in step]

# plot step function
%matplotlib inline
plt.plot(x,func,'r--',linewidth = 3)

```

```python
# create dataset based on this steppy
num_pts = 30
pts = np.random.permutation(x)[:num_pts]
vs_pts = [x[s] for s in vs[:-1]]
vs_pts.append(1)
pt_levs = []
for pt in pts:
    val = 0
    for v in range(len(vs_pts)):
        if pt > vs_pts[v] and pt <= vs_pts[v+1]:
            val = levels[v]
            pt_levs.append(val)
            break

pt_levs = [s[0] for s in pt_levs]
pt_noise = 0.09*np.random.randn(len(pt_levs),)
pt_levs+=pt_noise
plt.scatter(pts,pt_levs)
plt.plot(x,func,'r--',linewidth = 3)
```

```python
x = np.asarray(x)
x.shape = (len(x),1)
func = np.asarray(func)
func.shape = (len(func),1)
function = np.concatenate([x,func],axis = 1)
```

```python
f = pd.DataFrame(function,index = None)
f.to_csv('steppy_func.csv',header = None, index = None)
```
